# Часть II. Построение модели машинного обучения для набора данных: Деревья

## Задание

Вам нужно решить задачу классификации с помощью алгоритма

Целевая переменная, результат: **rings**

#### 25. Разделить выборку на признаки (Х) и результат (Y)

```python
X = df.drop('rings', axis=1)
Y = df['rings']

print("Размерность признаков (X):")
print(X.shape)
print("Размерность целевой переменной (Y):")
print(Y.shape)
print("Первые 5 строк X:")
print(X.head())
print("\nПервые 5 значений Y:")
print(Y.head())
```

#### 26. Разделить на обучающую и тестовую выборки

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print(f"Размер обучающей выборки X: {X_train.shape}")
print(f"Размер тестовой выборки X: {X_test.shape}")
print(f"Размер обучающей выборки Y: {Y_train.shape}")
print(f"Размер тестовой выборки Y: {Y_test.shape}")
```

#### 27. Сформировать модель машинного обучения

```python
from sklearn.tree import DecisionTreeClassifier

# Инициализация модели Дерева решений
model = DecisionTreeClassifier(random_state=42)

print("Модель машинного обучения (Decision Tree Classifier) сформирована.")
```

#### 28. Обучить модель

```python
model.fit(X_train, Y_train)
print("Модель успешно обучена.")
```

#### 29. Оценить качество модели

```python
from sklearn.metrics import accuracy_score, classification_report

# Предсказания на тестовой выборке
Y_pred = model.predict(X_test)

# Оценка точности (Accuracy)
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Точность модели на тестовой выборке: {accuracy:.4f}")

# Отчет о классификации (Precision, Recall, F1-score для каждого класса)
print("\nОтчет о классификации:")
print(classification_report(Y_test, Y_pred))
```
