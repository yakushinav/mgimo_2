#### 11. Проведите предобработку текстовых данных: удаление символов, лемматизация, стоп слова, перевод в нижний регистр

```python
import nltk
import re
import pandas as pd # Import pandas as it's used in pd.isna()

print("11. Предобработка текстовых данных:")

# Download necessary NLTK data
try:
    nltk.data.find('corpora/wordnet')
except LookupError:
    nltk.download('wordnet')
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Identify all columns in the DataFrame df that have an object data type.
object_columns = df.select_dtypes(include='object').columns

print(f"Обнаружены категориальные поля для предобработки: {list(object_columns)}")

# Define a preprocessing function
def preprocess_text(text):
    if pd.isna(text): # Handle NaN values
        return ""
    text = str(text)
    # a. Convert to lowercase
    text = text.lower()
    # b. Remove leading and trailing whitespace
    text = text.strip()
    # c. Remove any special characters, keeping only alphanumeric characters and spaces
    text = re.sub(r'[^a-z0-9\s]', '', text)
    # e. Tokenize the cleaned string into individual words.
    words = text.split()
    # f. Lemmatize each word
    words = [lemmatizer.lemmatize(word) for word in words]
    # g. Remove any stop words
    words = [word for word in words if word not in stop_words]
    # h. Join the remaining words back into a single string
    return ' '.join(words)

# Apply the preprocessing function to each object column
for column in object_columns:
    df[column] = df[column].apply(preprocess_text)

print("Предобработка текстовых полей завершена.")
print("\n" + "="*80 + "\n")
```

#### 12. Фильтрация. Выберите все строки таблицы, для которых значение поля num-of-cylinders больше 7

```python
print("12. Фильтрация: num-of-cylinders > 7")

def convert_cylinders(value):
    if isinstance(value, str):
        value = value.strip().lower()
        if value == 'two': return 2
        elif value == 'three': return 3
        elif value == 'four': return 4
        elif value == 'five': return 5
        elif value == 'six': return 6
        elif value == 'eight': return 8
        elif value == 'twelve': return 12
    try:
        return int(value)
    except (ValueError, TypeError):
        return None # Handle cases where conversion is not possible

df['num-of-cylinders_numeric'] = df['num-of-cylinders'].apply(convert_cylinders)

filtered_12 = df[df['num-of-cylinders_numeric'] > 7]
print(f"Найдено строк, где num-of-cylinders > 7: {len(filtered_12)}")
print("Первые 5 отфильтрованных строк:")
print(filtered_12.head())
print("\n" + "="*80 + "\n")
```
