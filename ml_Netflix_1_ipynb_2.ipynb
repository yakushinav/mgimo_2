#### 11. Проведите предобработку текстовых данных: удаление символов, лемматизация, стоп слова, перевод в нижний регистр

```python
# 11. Предобработка текстовых данных - ИСПРАВЛЕННЫЙ КОД (только английский)
import re
import nltk
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Скачиваем данные NLTK
nltk.download('stopwords', quiet=True)

# Инициализация для английского текста (Netflix данные на английском)
stop_words = set(stopwords.words('english'))
stemmer = SnowballStemmer('english')

def preprocess_text(text):
    if pd.isna(text):
        return ''
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Удаляем символы кроме букв
    words = text.split()
    words = [stemmer.stem(word) for word in words if word not in stop_words and len(word) > 2]
    return ' '.join(words)

# Применяем к текстовым полям (проверяем существование колонок)
text_columns = ['name', 'director', 'genre', 'description', 'country']
existing_text_cols = [col for col in text_columns if col in df.columns]
print(f"Обрабатываем колонки: {existing_text_cols}")

for col in existing_text_cols:
    df[col + '_clean'] = df[col].apply(preprocess_text)
    print(f"Обработана колонка: {col}")
```
